```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data = pd.read_csv("student.csv")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.describe()
```

```python
X=df['weekly_self_study_hours'].values.reshape(-1,1)
Y=df['math_score'].values.reshape(-1 , 1)
```

```python
X_train ,X_test , Y_train , Y_test = train_test_split(X ,Y ,  test_size = 0.3 , random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train , Y_train)
Y_pred = regressor.predict(X_test)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
plt.scatter(X_test ,Y_test)
plt.plot(X_test , Y_pred , color= 'red')
plt.show()
```

```python
print("Mse is:", metrics.mean_squared_error(Y_test, Y_pred))  
print("MAE is:", metrics.mean_absolute_error(Y_test, Y_pred))  
print("Rmse is :",np.sqrt(metrics.mean_squared_error(Y_test , Y_pred)))
print("R2score :" , metrics.r2_score(Y_test , Y_pred)*100)
```
